```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
# Load an image
image = cv2.imread("/content/thumbnail-removebg-preview.png")  # Replace with your image path
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)  # Convert BGR to RGB

# Show original image
plt.imshow(image)
plt.axis("off")
plt.title("Original Image")
plt.show()
```

```python
# Reshape the image to a 2D array of pixels
pixels = image.reshape((-1, 3))  # Shape becomes (total_pixels, 3)
```

```python
# Define number of colors (K value)
K = 64 # Try different values like 8, 16, 32

# Apply K-Means clustering
kmeans = KMeans(n_clusters=K, random_state=42, n_init=10)
kmeans.fit(pixels)

# Replace each pixel with its nearest cluster center
compressed_pixels = kmeans.cluster_centers_[kmeans.labels_]
compressed_pixels = np.clip(compressed_pixels.astype('uint8'), 0, 255)

# Reshape back to original image shape
compressed_image = compressed_pixels.reshape(image.shape)
```

```python
# Show compressed image
plt.imshow(compressed_image)
plt.axis("off")
plt.title(f"Compressed Image with {K} Colors")
plt.show()
```
